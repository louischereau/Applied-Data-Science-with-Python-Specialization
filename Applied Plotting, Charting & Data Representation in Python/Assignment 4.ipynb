```python
url = 'https://en.wikipedia.org/wiki/Michigan_Wolverines'

olymp = pd.read_html(url,header=0)[9].dropna(how='all',axis=1)
olymp.loc[olymp['First'] == 'Meryl/Charlie', 'First'] = 'Meryl'
olymp.loc[olymp['Last'] == 'Davis/White', 'Last'] = 'Davis'

mc = olymp.loc[olymp['First'] == 'Meryl/Charlie']
mc.loc[:,'First'] = 'Charlie'
mc.loc[:,'Last'] = 'White'

olymp = pd.concat([olymp,mc]).sort_values('Year')

males = []
females = []

for name in olymp['First']:
    if name.endswith(('ey', 'fer', 'yn','a', 'i', 'se', 'ice','ish','oan','ny','ate','ones','yl')):
        females.append(name)
    else:
        males.append(name)

olymp_males = olymp.loc[olymp['First'].isin(males)].sort_values('Year').drop(columns = {'First', 'Last', 'Sport', 'Event', 'Country'})
olymp_females = olymp.loc[olymp['First'].isin(females)].sort_values('Year').drop(columns = {'First', 'Last','Sport', 'Event', 'Country'})

olymp_males['Bronze'] = olymp_males['Medal'] == 'bronze'
olymp_males['Silver'] = olymp_males['Medal'] == 'silver'
olymp_males['Gold'] = olymp_males['Medal'] == 'gold'

olymp_females['Bronze'] = olymp_females['Medal'] == 'bronze'
olymp_females['Silver'] = olymp_females['Medal'] == 'silver'
olymp_females['Gold'] = olymp_females['Medal'] == 'gold'

olymp_m = olymp_males.drop(columns = 'Medal').astype(int).groupby('Year').sum()
olymp_f= olymp_females.drop(columns = 'Medal').astype(int).groupby('Year').sum()

olymp_men = olymp_males.drop(columns = 'Medal').astype(int).groupby('Year').count().drop(columns = {'Silver', 'Gold'}).rename(columns = {'Bronze':'Medals'})
olymp_women = olymp_females.drop(columns = 'Medal').astype(int).groupby('Year').count().drop(columns = {'Silver', 'Gold'}).rename(columns = {'Bronze':'Medals'})

olymp = olymp_m.merge(olymp_f, how = 'outer', left_index = True, right_index = True).fillna(0).agg(np.sum)

olymp2 = olymp_men.merge(olymp_women, how = 'outer', left_index = True, right_index = True).fillna(0)

olymp_total = olymp2.agg(np.sum)

olymp_mean = olymp2.agg(np.mean)

olymp_mean

```
